# Comparing multiple PyWGCNAs
PyWGCNA also supports comparing more than two PyWGCNA objects to each other to find out how reproducible modules are across runs.
First, read your PyWGCNa object using the `readWGCNA()` function.

For this tutorial, I used three datasets from three different mouse models:
1. 5xFAD mouse model: [Systematic phenotyping and characterization of the 5xFAD mouse model of Alzheimer’s disease](https://www.nature.com/articles/s41597-021-01054-y)
2. 3xTg-AD mouse model:[Systematic Phenotyping and Characterization of the 3xTg-AD Mouse Model of Alzheimer’s Disease](https://www.frontiersin.org/articles/10.3389/fnins.2021.785276/full)
3. Trem2: [A Trem2*R47H mouse model without cryptic splicing drives age- and disease-dependent tissue damage and synaptic loss in response to plaques](https://www.biorxiv.org/content/10.1101/2022.03.09.483490v1.abstract)

I applied PyWGCNA with default parameters in each of them seperately and then compared their modules

## Read all of your PyWGCNA objects

```python
import PyWGCNA
pyWGCNA_5xFAD = PyWGCNA.readWGCNA("5xFAD.p")
pyWGCNA_3xTgAD = PyWGCNA.readWGCNA("3xTgAD_paper/3xTgAD.p")
pyWGCNA_Trem2 = PyWGCNA.readWGCNA("modelAD_TREM2_5xFAD_BL6_4mon_12mon.p")
pyWGCNA_Trem2.name = "TREM2"
```

## Comparing two or more PyWGCNA objects
After reading the objects you only need to call `compareNetworks()` function on your two PyWGCNA, the results of the comparison are stored in three tables:
1. Jaccard similarity matrix which reports the similarity between each module across networks
2. Fraction matrix which reports the fraction of common genes between each module across networks
3. P-value matrix which reports the significance of the similarity between each set of modules across the networks

```python
comparison = PyWGCNA.compareNetworks(PyWGCNAs = [pyWGCNA_5xFAD, 
                                                 pyWGCNA_3xTgAD,
                                                 pyWGCNA_Trem2])
```

### Jacard similarity matrix

```python
comparison.jaccard_similarity.head(5)
```

### Fraction matrix

```python
comparison.fraction.head(5)
```

### P-value matrix

```python
comparison.P_value.head(5)
```

## Plotting results
There are two options to plot the results:
1. Display the Jaccard similarity matrix as a graph by using `plot_jaccard_similarity()` function.
2. Display all three matrices at once using `plotHeatmapComparison()` function.

```python
color = {"5xFAD": "lightblue", 
         "3xTgAD": "lightgreen", 
         "TREM2": "lightpink"}
comparison.plotJaccardSimilarity(color=color,
                                   cutoff=0.2,
                                   plot_format="pdf",
                                   file_name="jaccard_similarity_5xFAD_3xTgAD_Trem2")
```

```python
comparison.plotHeatmapComparison(plot_format="pdf",
                                 file_name="heatmap_comparison_5xFAD_3xTgAD_Trem2")
```

```python
color = {"5xFAD": "lightblue", 
         "3xTgAD": "lightgreen", 
         "TREM2": "lightpink"}
comparison.plotBubbleComparison(color=color,
                                plot_format="pdf",
                                file_name="bubble_comparison_5xFAD_3xTgAD_Trem2")
```

## Saving and loading your comparison object
You can save or load your comparison object with `saveComparison()` or `readComparison()` functions respectively.

```python
comparison.saveComparison(name="comparison_5xFAD_3xTgAD_Trem2")
```

```python
import PyWGCNA
comparison = PyWGCNA.readComparison('comparison_5xFAD_3xTgAD_Trem2.p')
```
