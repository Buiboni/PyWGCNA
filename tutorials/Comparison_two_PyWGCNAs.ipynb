# Comparing two PyWGCNAs
you can also compare to pyWGCNA objects to  each other to find out how each group of cluster is correlation with each others.
For achieving that, first you need to read your PyWGCNa object by `readWGCNA()` function.

For this tutorial, I used two datasets from two mouse models:
1. 5xFAD mouse model: [Systematic phenotyping and characterization of the 5xFAD mouse model of Alzheimer’s disease](https://www.nature.com/articles/s41597-021-01054-y)
2. 3xTg-AD mouse model:[Systematic Phenotyping and Characterization of the 3xTg-AD Mouse Model of Alzheimer’s Disease](https://www.frontiersin.org/articles/10.3389/fnins.2021.785276/full)

## Read all of your PyWGCNA objects

```python
import PyWGCNA
pyWGCNA_5xFAD = PyWGCNA.readWGCNA("5xFAD.p")
pyWGCNA_3xTgAD = PyWGCNA.readWGCNA("3xTgAD_paper/3xTgAD.p")
```

## Comparing two or more PyWGCNA objects
After reading the objects you only need to call `compareNetworks()` function on your two PyWGCNA, you can find the comparison results in three tables:
1. jacard similarity matrix which show you the similarity between each modules across networks
2. fraction matrix which show you the fraction of common genes between each modules across networks
3. p-value matrix which show you how much the similarity is significant between each modules across networks

```python
comparison = PyWGCNA.compareNetworks(PyWGCNAs = [pyWGCNA_5xFAD, pyWGCNA_3xTgAD])
```

### Jacard similarity matrix

```python
comparison.jaccard_similarity.head(5)
```

### Fraction matrix

```python
comparison.fraction.head(5)
```

### P-value matrix

```python
comparison.P_value.head(5)
```

## Ploting comparison
THere is two options to plot the results:
1. You can plot the jaccard similarity matrix as a graph by using `plot_jaccard_similarity()` function.
2. plot all three matrix as once using `plotHeatmapComparison()` function.

```python
color = {"5xFAD": "lightblue", 
         "3xTgAD": "lightgreen"}
comparison.plotJaccardSimilarity(color=color,
                                   cutoff=0.1,
                                   plot_format="pdf",
                                   file_name="jaccard_similarity_5xFAD_3xTgAD")
```

```python
comparison.plotHeatmapComparison(plot_format="pdf",
                                 file_name="heatmap_comparison_5xFAD_3xTgAD")
```

```python
color = {"5xFAD": "lightblue", 
         "3xTgAD": "lightgreen"}
comparison.plotBubbleComparison(color=color,
                                plot_format="pdf",
                                file_name="bubble_comparison_5xFAD_3xTgAD")
```

## Saving and loading your comparison object
you can save or load your comparison object with saveComparison() or readComparison() function.

```python
comparison.saveComparison(name="comparison_5xFAD_3xTgAD")
```

```python
import PyWGCNA
comparison = PyWGCNA.readComparison('comparison_5xFAD_3xTgAD.p')
```
